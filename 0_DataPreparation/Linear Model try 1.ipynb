```python
import statsmodels.formula.api as smf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
from pathlib import Path
base = Path.cwd()  # oder Path("/workspaces/datascience-project-new")
train_data = pd.read_csv('data/train_data.csv')


# Fit the linear model
mod = smf.ols('Umsatz ~ C(Warengruppe)', data=train_data).fit()

# Print the summary
print(mod.summary())
```

```python
# Trainingsdaten laden
train_data = pd.read_csv('data/train_data.csv')

# Modell trainieren
mod = smf.ols('Umsatz ~ C(Warengruppe)', data=train_data).fit()
print(mod.summary())

# Test-/Vorhersagedaten laden (die 1830 Zeilen ohne Umsatz)
test_data = pd.read_csv('data/test_data.csv')  

# Vorhersagen machen
test_data['Umsatz_Vorhersage'] = mod.predict(test_data)

# Nur ID und Umsatz_Vorhersage speichern
predictions_output = test_data[['id', 'Umsatz_Vorhersage']]
predictions_output.to_csv('data/predictions.csv', index=False)

# Erste Vorhersagen anzeigen
print("\nErste 10 Vorhersagen:")
print(predictions_output.head(10))
```
