```python
# Modell trainieren ohne Windgeschwindigkeit und Fastzeit mit Daten von df_final_new.csv
# FeiertagSilvester FeiertagHeiligabend Datum WettercodeDunst Fastenzeit

mod = smf.ols('Umsatz ~ + Temperatur + KielerWoche + WettercodeExtremwetter + WettercodeKlar + WettercodeNachSchlechtemWetter + WettercodeNebel + WettercodeRegen + WettercodeSchnee + WarengruppeBread + WarengruppeRolls + WarengruppeConfectionery + WarengruppeCake + WarengruppeSeasonalBread + WochentagFreitag + WochentagSamstag + WochentagSonntag + WochentagUnterDerWoche + Januar + Maerz + April + Mai + Juni + Juli + August + September + Oktober + November + Dezember + BewoelkungWolkenBedeckt + BewoelkungWolkenBewoelkt + BewoelkungWolkenKlar + FeiertagChristiHimmelfahrt + FeiertagErsterMai + FeiertagPfingstmontag + FeiertagTagDerDeutschenEinheit',
               data=train_data).fit()
print(mod.summary())

validation_data['Umsatz_Pred'] = mod.predict(validation_data)

rmse_val = np.sqrt(mean_squared_error(
    validation_data['Umsatz'],
    validation_data['Umsatz_Pred']
))

print(f"Validation RMSE: {rmse_val:.2f}")

# Test-/Vorhersagedaten laden (die 1830 Zeilen ohne Umsatz)
test_data = pd.read_csv('../0_DataPreparation/data/test_data.csv')  

# Vorhersagen machen
test_data['Umsatz_Vorhersage'] = mod.predict(test_data)

# Nur ID und Umsatz_Vorhersage speichern
predictions_output = test_data[['id', 'Umsatz_Vorhersage']]
predictions_output.to_csv('../0_DataPreparation/data/predictions_preencoded_imputated.csv', index=False)


# Erste Vorhersagen anzeigen
print("\nErste 10 Vorhersagen:")
print(predictions_output.head(10))
```
